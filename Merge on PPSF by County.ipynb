# Notes
- Merge two dataframes by the FIPS
- Zillow csv problem: Some zillow files work in pd.read_csv(file) with no problem but if it doesn't work, try pd.read_csv(file, encoding="latin")

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
county_stats_1_file = "raw data/2018_all_county_data_1.csv"
stats_1_df = pd.read_csv(county_stats_1_file, encoding="latin")
stats_1_df.head(10)
```

```python
top_20_ppsf_file= "Zillow CSV results/top_counties_ppsf_1.csv"
top_20_ppf_df=pd.read_csv(top_20_ppsf_file, encoding="latin")
top_20_ppf_df.head(10)
```

```python
top_ppsf_merge1_df=pd.merge(top_20_ppf_df,stats_1_df,how='left', on=['FIPS'])
#top__ppsf_merge1_df.head()
#print(len(top_ppsf_merge1_df))
```

```python
top_ppsf_merge1_df
```

```python
bottom_20_ppsf_file= "Zillow CSV results/bottom_counties_ppsf_1.csv"
bottom_20_ppsf_df=pd.read_csv(bottom_20_ppsf_file, encoding="latin")
bottom_20_ppsf_df.head()
```

```python
bottom_ppsf_merge1_df=pd.merge(bottom_20_ppf_df,stats_1_df,how='left', on=['FIPS'])
bottom_ppsf_merge1_df
```
